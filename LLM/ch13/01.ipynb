seq2seq
```
영어문장을 한국어로 번역, 질의 --> 답변생성, 문장요약등.
why?
    입력과 출력의 길이가 다를 수 있는 문제(번역, 요약, 챗봇)
    고정크기의 피처벡터는 순서정보와 문맥관계를 충분히 반영 못함
    단순 분류모델은 시퀀스 간 종속적 생성에 부적합
```
동작원리
```
    1. Encoder : 입력토큰들을 순차적으로 처리 -> 마지막/전체 Hidden state 집약
    2. Context(요약표현) : Encoder의 출력을 기반으로 입력시퀀스의 의미를 압축
    3. Decoder : Context를 초기상태로 사용 -> 출력토큰들을 순차적으로 생성
    4. 종료 : 특정종료토큰 도달 시 종료
```
